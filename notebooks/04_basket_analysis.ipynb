```python
def categorias_diferentes(itemset):
    subcategoria_a_categoria = crudos[['subcategoria', 'categoria']].drop_duplicates().set_index('subcategoria')['categoria'].to_dict()
    # Convertir el itemset en un conjunto (frozenset) en caso de que sea lista o similar
    if isinstance(itemset, list):
        itemset = frozenset(itemset)
    
    # Obtener las categorías correspondientes a cada subcategoría
    categorias = {subcategoria_a_categoria[sub] for sub in itemset if sub in subcategoria_a_categoria}
    
    # Retornar True si hay más de una categoría, lo que significa que las subcategorías son de diferentes categorías
    return len(categorias) > 1
```

```python
subcategoria_a_categoria = crudos[['subcategoria', 'categoria']].drop_duplicates().set_index('subcategoria')['categoria'].to_dict()
subcategoria_a_familia = crudos[['subcategoria', 'familia']].drop_duplicates().set_index('subcategoria')['familia'].to_dict()
```

```python
def familias_diferentes(itemset):
    subcategoria_a_familia = crudos[['subcategoria', 'familia']].drop_duplicates().set_index('subcategoria')['familia'].to_dict()
    # Convertir el itemset en un conjunto (frozenset) en caso de que sea lista o similar
    if isinstance(itemset, list):
        itemset = frozenset(itemset)
    
    # Obtener las categorías correspondientes a cada subcategoría
    familias = {subcategoria_a_familia[sub] for sub in itemset if sub in subcategoria_a_familia}
    
    # Retornar True si hay más de una categoría, lo que significa que las subcategorías son de diferentes categorías
    return len(familias) > 1
```

```python
def barras_basket_cat(datos,cantidad, multiples = False, cross = False):  
    # Ordenar los valores de soporte de mayor a menor
    df = datos.sort_values(by='support', ascending=False)
    if multiples and cross:
        df = df[df['itemsets'].apply(lambda x: len(x) > 1)]
        df = df[df['itemsets'].apply(categorias_diferentes)].sort_values(by='support',ascending = False)
    elif multiples:
        df = df[df['itemsets'].apply(lambda x: len(x) > 1)]
    elif cross:
        df = df[df['itemsets'].apply(    categorias_diferentes)].sort_values(by='support',ascending = False)
    # Seleccionar los n más grandes (en este ejemplo, mostramos los 3 más grandes)
    top_n_df = df.head(cantidad)
    top_n_df['itemsets'] = top_n_df['itemsets'].apply(lambda x: ', '.join(list(x)))
    # Crear el gráfico de barras
    plt.figure(figsize=(15, 6))
    bars = plt.bar(range(len(top_n_df)), top_n_df['support'], tick_label=[i for i in top_n_df['itemsets']],color='skyblue')
    
    # Añadir los valores de soporte centrados en las barras
    for bar, support in zip(bars, top_n_df['support']):
        plt.text(
            bar.get_x() + bar.get_width() / 2,  # Posición X
            bar.get_height() / 2,               # Posición Y (mitad de la barra)
            f'{support:.3f}',                   # Formato del texto (cuatro decimales)
            ha='center', va='center',           # Centrar horizontal y verticalmente
            color='black', fontsize=8         # Color del texto y tamaño de la fuente
        )
    
    # Configuraciones adicionales
    plt.title(f'Top {cantidad} Conjuntos Subcategorias Frecuentes con Mayor Soporte')
    plt.ylabel('Soporte')
    plt.xlabel('Conjuntos de Subcategorías')
    plt.xticks(rotation=90)
    plt.tight_layout()
    
    # Mostrar el gráfico
    plt.show()
```

```python
def barras_basket_fam(datos,cantidad, multiples = False, cross = False):  
    # Ordenar los valores de soporte de mayor a menor
    df = datos.sort_values(by='support', ascending=False)
    if multiples and cross:
        df = df[df['itemsets'].apply(lambda x: len(x) > 1)]
        df = df[df['itemsets'].apply(familias_diferentes)].sort_values(by='support',ascending = False)
    elif multiples:
        df = df[df['itemsets'].apply(lambda x: len(x) > 1)]
    elif cross:
        df = df[df['itemsets'].apply(familias_diferentes)].sort_values(by='support',ascending = False)
    # Seleccionar los n más grandes (en este ejemplo, mostramos los 3 más grandes)
    top_n_df = df.head(cantidad)
    top_n_df['itemsets'] = top_n_df['itemsets'].apply(lambda x: ', '.join(list(x)))
    # Crear el gráfico de barras
    plt.figure(figsize=(15, 6))
    bars = plt.bar(range(len(top_n_df)), top_n_df['support'], tick_label=[i for i in top_n_df['itemsets']],color='skyblue')
    
    # Añadir los valores de soporte centrados en las barras
    for bar, support in zip(bars, top_n_df['support']):
        plt.text(
            bar.get_x() + bar.get_width() / 2,  # Posición X
            bar.get_height() / 2,               # Posición Y (mitad de la barra)
            f'{support:.3f}',                   # Formato del texto (cuatro decimales)
            ha='center', va='center',           # Centrar horizontal y verticalmente
            color='black', fontsize=8         # Color del texto y tamaño de la fuente
        )
    
    # Configuraciones adicionales
    plt.title(f'Top {cantidad} Conjuntos Subcategorias Frecuentes con Mayor Soporte')
    plt.ylabel('Soporte')
    plt.xlabel('Conjuntos de Subcategorías')
    plt.xticks(rotation=90)
    plt.tight_layout()
    
    # Mostrar el gráfico
    plt.show()
```

```python
# Función para verificar que los antecedentes y los consecuentes no pertenezcan a la misma categoría
def categorias_diferentes(antecedents, consequents):
    categorias_antecedents = {subcategoria_a_categoria[sub] for sub in antecedents if sub in subcategoria_a_categoria}
    categorias_consequents = {subcategoria_a_categoria[sub] for sub in consequents if sub in subcategoria_a_categoria}
    # Verificar si las categorías de antecedentes y consecuentes son diferentes
    return categorias_antecedents.isdisjoint(categorias_consequents)  # Retorna True si no hay categorías comunes

# Función para verificar que los antecedentes y los consecuentes no pertenezcan a la misma familia
def familias_diferentes(antecedents, consequents):
    familias_antecedents = {subcategoria_a_familia[sub] for sub in antecedents if sub in subcategoria_a_familia}
    familias_consequents = {subcategoria_a_familia[sub] for sub in consequents if sub in subcategoria_a_familia}
    # Verificar si las familias de antecedentes y consecuentes son diferentes
    return familias_antecedents.isdisjoint(familias_consequents)  # Retorna True si no hay familias comunes
```
